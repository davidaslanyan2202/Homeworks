```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lung_cancer_df = pd.read_csv("lung_cancer_prediction_dataset.csv")
air_pollution_df = pd.read_csv("global_air_pollution_dataset.csv")
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=lung_cancer_df["Annual_Lung_Cancer_Deaths"], color="lightcoral")
plt.ylabel("Annual Lung Cancer Deaths")
plt.title("Boxplot of Lung Cancer Deaths Distribution")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(air_pollution_df["PM2.5_AQI_Value"].dropna(), bins=30, color="darkblue", edgecolor="black", alpha=0.7)
plt.xlabel("PM2.5 AQI Value")
plt.ylabel("Frequency")
plt.title("Histogram of PM2.5 AQI Values")
plt.grid()
plt.show()
```

```python
mortality_rate_filtered = pd.to_numeric(lung_cancer_df["Mortality_Rate"], errors="coerce").dropna()
mortality_rate_filtered = mortality_rate_filtered[mortality_rate_filtered > 0]  # Remove zero values to avoid KDE issues

plt.figure(figsize=(6, 4))
sns.histplot(mortality_rate_filtered, bins=30, kde=True, color="purple", alpha=0.6)
plt.xlabel("Mortality Rate")
plt.ylabel("Density")
plt.title("Histogram with KDE Overlay for Lung Cancer Mortality Rate")
plt.grid()
plt.show()
```

```python
np.random.seed(42)  
normal_values = np.random.normal(loc=0, scale=1, size=100)
logistic_values = np.random.logistic(loc=0, scale=1, size=100)

plt.figure(figsize=(6, 4))
plt.scatter(normal_values, logistic_values, color="brown", alpha=0.7)
plt.xlabel("Normal Distribution Values")
plt.ylabel("Logistic Distribution Values")
plt.title("Scatter Plot of Normal vs. Logistic Distribution")
plt.grid()
plt.show()
```
